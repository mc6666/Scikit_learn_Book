```python
from seaborn import load_dataset
df = load_dataset('titanic')
df.head()
```

```python
df.embarked = df.embarked.fillna(method='ffill')
df.head()
```

```python
from CHAID import Tree

independent_variable_columns = ['sex', 'embarked']
dep_variable = 'survived'
tree = Tree.from_pandas_df(df, dict(zip(independent_variable_columns, ['nominal'] * 3)), dep_variable
          ,dep_variable_type='categorical', max_depth=5, min_parent_node_size=2, alpha_merge=0.1) 
tree.print_tree()
```

```python
from CHAID import Tree

independent_variable_columns = ['fare', 'sex', 'embarked']
dep_variable = 'survived'
tree = Tree.from_pandas_df(df, dict(zip(independent_variable_columns, ['ordinal'] + ['nominal'] * 3)), dep_variable
          ,dep_variable_type='categorical', max_depth=5, min_parent_node_size=2, alpha_merge=0.1) 
tree.print_tree()
```
